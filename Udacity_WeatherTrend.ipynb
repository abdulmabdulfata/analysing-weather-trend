```python
#Importing the required libraries for the project

import nbconvert
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Importing the two .csv files for global and Maiduguri - Nigeria
data_global = pd.read_csv('data_global.csv')
data_maiduguri = pd.read_csv('data_maiduguri.csv')
```

Both the data are okay. Next is to plot the temprature against year for the global weather data and the Maiduguri - Nigeria the city closest to me.

```python
#Plotting the graph of global weather data, temprature against year.
data_global.iloc[156:].plot(kind='line', x='year', y='avg_temp', color='blue')
plt.show()
```

```python
#Plotting the graph of Maiduguri - Nigeria weather data, temprature against year.
data_maiduguri.plot(kind='line', x='year', y='avg_temp', color='black')
plt.show()
```

To smooth the two plotted grapha of global weather and Maiduguri or observe long term trends more easily, moving average tempratures will be plotted against years.

```python
#Plotting the two lines on the same chart together, moving avarage temperature against years for the global and Maiduguri - Nigeria.
data_global['temperature_moving_average'] = data_global.iloc[:, 1].rolling(window=14).mean()
data_maiduguri['temperature_moving_average'] = data_maiduguri.iloc[:, 4].rolling(window=14).mean()

plt.figure(figsize=(20,10))
plt.plot(data_global['year'], data_global['avg_temp'], label= 'Global Weather Trend', color = 'blue')
plt.plot(data_maiduguri['year'], data_maiduguri['avg_temp'], label='Maiduguri Weather Trend', color = 'black')
plt.xlabel('Year')
plt.ylabel('Temperature Moving Average')
plt.legend()
plt.title('Temperatures Against Year')
plt.show()
```

- Using pandas library rolling function 4 moving year was calculated.
- As earlier used, matpotlib was used in plotting 14 moving averages for the weather trends for both the global Maiduguri - Nigeria.

```python
#Plotting the global average temprature
data_global.iloc[120:].plot(kind='line', x='year', y='temperature_moving_average', color='blue', title='Global Average Temperature')

#Plotting the Maiduguri - Nigeria average temprature
data_maiduguri.plot(kind='line', x='year', y='temperature_moving_average', color='black', title='Maiduguri Average Temperature')
```

The analysis for the weather trend is as follows:
- The global weather have been increasing upwards for the past 200 years. During late 1800 and world war II there was an inflation.
- Local temperatures trend in Maiduguri - Nigeria have trended about 19-20 degrees celsius above the global average weather trend.
- Both the temperatures have increased almost one degree celsius/century.
- From 1980 to date, both the global and Maiduguri-Nigeria average weather have been trending the same direction (becoming hotter).
